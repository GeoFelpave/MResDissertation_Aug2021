### **Plotting data using a shapefile as a mask for the UK**

**GPM-IMERG example**

```python
# Importing mask modules
import geopandas
import rioxarray
from shapely.geometry import mapping

# Reading the data
ds1 = xr.open_dataset(output_folder + 'test_daily_pr_intensity.nc', decode_timedelta=False)

# Setting spatial dimmension in nc data
ds1.rio.set_spatial_dims(x_dim="lon", y_dim="lat", inplace=True)
ds1.rio.write_crs("epsg:4326", inplace=True)

# Set mask based on boundary
uk_admn = geopandas.read_file('/mnt/d/MRes_dataset/active_Data/101_admin/uk_admin_boundary_py_nasa_pp_countryOutlineFromGiovanni.shp', crs="epsg:4326")
# Data for UK
gpm_ToClip = ds1
uk_ssdi = gpm_ToClip.rio.clip(uk_admn.geometry.apply(mapping), uk_admn.crs, drop=False)

# Plotting UK data

# change date from object to datetime64[ns]
datetimeindex = uk_ssdi.indexes["time"].to_datetimeindex()
uk_ssdi["time"] = datetimeindex

# create plots
fig, (ssdi2000, ssdi2001) = plt.subplots(1, 2, figsize=(10, 5))
(
    uk_ssdi['ssdi']
    .sel(time=slice("2000-01-01", "2000-01-01"))
).plot(ax=ssdi2000, x='lon', y='lat')
(
    uk_ssdi['ssdi']
    .sel(time=slice("2001-01-01", "2001-01-01"))
).plot(ax=ssdi2001, x='lon', y='lat')

ds1.close()
```

**HadUK-Grid Example**

```python
# Reading the data
ds1 = xr.open_dataset(output_folder + 'test_daily_pr_intensity_HadUK.nc', decode_timedelta=False)

# Plotting UK data
# create plots
fig, (ssdi2000, ssdi2001) = plt.subplots(1, 2, figsize=(10, 5))
(
    ds1['ssdi']
    .sel(time=slice("2000-01-01", "2000-01-01"))
).plot(ax=ssdi2000, x='longitude', y='latitude')
ssdi2000.set_xlim([-9, 2])
ssdi2000.set_ylim([48, 61])

(
    ds1['ssdi']
    .sel(time=slice("2001-01-01", "2001-01-01"))
).plot(ax=ssdi2001, x='longitude', y='latitude')
ssdi2001.set_xlim([-9, 2])
ssdi2001.set_ylim([48, 61])


ds1.close()
```
