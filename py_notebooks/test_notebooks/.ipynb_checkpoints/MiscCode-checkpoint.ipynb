**Create Temperature Variables**

```python
# Open file as nc to add new information
# mode can be 'w', 'r', 'a' or 'r+'
# ncfile = netCDF4.Dataset('in.nc',mode='',format='NETCDF4_CLASSIC')
dataset_nc = ncd.Dataset(FILEPATH + '/daily_imerg_precipitation_2001.nc','a')

# create fake data
print(dataset_nc.dimensions) # This provides the information for the size of the data array
data_arr_max = np.random.uniform(low=0,high=330,size=(365,111,121))
data_arr_min = np.random.uniform(low=0,high=330,size=(365,111,121))

# Define temperature variable to hold the fake data
temp_max = dataset_nc.createVariable('tmax',np.float32,('time', 'lon', 'lat')) # note: unlimited dimension is leftmost
temp_max.units = 'K' # degrees Kelvin
temp_max.standard_name = 'fake_temperature_max' # this is a CF standard name

temp_min = dataset_nc.createVariable('tmin',np.float32,('time', 'lon', 'lat')) # note: unlimited dimension is leftmost
temp_min.units = 'K' # degrees Kelvin
temp_min.standard_name = 'fake_temperature_min' # this is a CF standard

# add fake data
temp_max[:,:,:] = data_arr_max
temp_min[:,:,:] = data_arr_min

# close the nc file with the changes.
dataset_nc.close(); print('Dataset is closed!')
```

```python
# Open the new file and save to avoid any future locks
dataset = xr.load_dataset(FILEPATH + '/daily_imerg_precipitation_2001.nc')
dataset.to_netcdf(FILEPATH + 'daily_imerg_precipitation_2001_fix.nc')
```

```python
# odifying the time variable to match datetime64 on the new file
dataset = xr.load_dataset(FILEPATH + '/daily_imerg_precipitation_2001_fix.nc')

# change date from object to datetime64[ns]
datetimeindex = dataset.indexes["time"].to_datetimeindex()
dataset["time"] = datetimeindex


# adding attributes for variable
dataset["time"].attrs['standard_name'] = 'time'
dataset["time"].attrs['bounds'] = 'time_bnds'
dataset["time"].attrs['origname'] = 'time'
dataset["time"].attrs['fullnamepath'] = '/time'

dataset.to_netcdf(FILEPATH + '/daily_imerg_precipitation_2001_final.nc')
```
